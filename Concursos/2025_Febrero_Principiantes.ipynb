```python
def vida_migrante(distancia:int, esperanza:int) -> int:
    return distancia*esperanza

distancia, esperanza = input().split()
print(vida_migrante(int(distancia), int(esperanza)))
```

## B. Conteo de Mariposas (V0)

image.png

```python
def conteo_mariposas(racimos:int, mariposas:int) -> int:
    return racimos*mariposas

racimos, mariposas = input().split()
print(conteo_mariposas(int(racimos), int(mariposas)))
```

## C. Migracin de Golondrinas en Amrica

image.png

image.png

```python
def golondrinas(cadena:str):
    cant, paradas = list(map(int, cadena.split()))

    for _ in range(paradas):
        mas, menos = list(map(int, input().split()))
        cant = cant + mas - menos

    return cant

cadena = input()
print(golondrinas(cadena))
```

## D. Migracin Anual de Aves

image.png

image.png

```python
def migracion(cadena:str):
    paradas = list(map(int, cadena.split()))
    return sum(paradas) + ((len(paradas)-1)*2)

print(migracion(input()))
```

## E. Migracin favorable

image.png

image.png

```python
def migracion_favorable(cadena:str):
    parada = list(cadena.split())
    paradas = "".join(parada)

    for _ in range(len(parada), 0, -1):
        if '1'*_ in paradas:
            return _
    return 0

input()
print(migracion_favorable(input()))
```

```python
def ecosistema(dep, X):
    cont = 0
    for _ in dep:
        if _[1]==X: cont+=1
    return cont

N, M, X = map(int, input().split())
dep = []
for _ in range(M): dep.append(list(map(int, input().split())))

print(ecosistema(dep, X))
```
